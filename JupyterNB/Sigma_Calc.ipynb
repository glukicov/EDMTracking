```python
from math import sqrt, radians, atan2, atan, tan, degrees
from IPython.display import Latex
import numpy as np
font_size=14
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import os, sys
sys.path.append('../CommonUtils/')
import CommonUtils as cu
```

```python
A_bz=     np.array([17.3,    7.8,       6.6,      3.1, -1.0])
A_bz_e=   np.array([11.8,    4.6,       4.0,     3.4 ,  3.4])
sigma=    np.array([24.221, 17.994,  14.789,  11.904, 8.139])
sigma_e=  np.array([6*1e-3, 2*1e-3,  2*1e-3,  2*1e-3, 2*1e-3])
                    
cuts = [r"$0<p<800$ MeV", r"$800<p<1200$ MeV", r"$1200<p<1500$ MeV", r"$1500<p<2000$ MeV", r"$2000<p<3100$ MeV"]
    
x=[1, 2, 3, 4, 5]
```

```python
fig, ax = cu.plot(x, A_bz, y_err=A_bz_e, c="g", marker="o", error=True, label="Run-1 S1218")
ax.set_xticks(x)
ax.set_xticklabels(cuts)
ax.set_ylabel(r"$A_{B_z} \ [\rm{\mu}$rad]")
# ax.set_xlim(0.7, 4.3)
# ax.set_ylim(-4, 30)
ax.set_xlabel("")
for tick in ax.get_xticklabels():
    tick.set_rotation(15)
plt.xticks(fontsize=14)
plt.tight_layout()
plt.legend(fontsize=14, loc=(0.55,0.85))
ax.tick_params(axis='x', which='minor', bottom=False, top=False)
fig.savefig("../fig/sum_mom_A_bz.png", dpi=300, bbox_inches='tight');
```

```python
fig, ax = cu.plot(x, sigma, y_err=sigma_e, c="g", marker="o", error=True, label="Run-1 S1218 \n"+r"$\Delta\sigma_{\theta_y}$"+r"$\sim10^{-3}$ mrad")
ax.set_xticks(x)
ax.set_xticklabels(cuts)
ax.set_ylabel(r"$\sigma_{\theta_y} \ [\rm{\mu}$rad]")
# ax.set_xlim(0.7, 4.3)
# ax.set_ylim(-4, 30)
ax.set_xlabel("")
for tick in ax.get_xticklabels():
    tick.set_rotation(15)
plt.xticks(fontsize=14)
plt.tight_layout()
plt.legend(fontsize=14, loc=(0.5,0.70))
ax.tick_params(axis='x', which='minor', bottom=False, top=False)
fig.savefig("../fig/sum_mom_Sigma.png", dpi=300, bbox_inches='tight');
```
